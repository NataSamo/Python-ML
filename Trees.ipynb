```python
Price = train['price_range'].squeeze()
train = train.drop(['price_range'], axis = 1)
train
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, Price, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state=8, max_depth=3)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

y_pred = clf.predict(X_test)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, display_labels=clf.classes_);
```

```python
from sklearn import tree

plt.figure(figsize=(20, 10))
tree.plot_tree(clf)
plt.show()
```

```python
clf.feature_importances_
```

```python
clf.feature_names_in_
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
scaler = preprocessing.MinMaxScaler()
scaler.fit(X_train)
X_train_norm = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)
X_test_norm = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
clf = DecisionTreeClassifier(random_state=8, max_depth=3)
clf.fit(X_train_norm, y_train)
clf.score(X_test_norm, y_test)
```

```python
y_pred = clf.predict(X_test_norm)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, display_labels=clf.classes_);
```
