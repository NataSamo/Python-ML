```python
from sklearn.preprocessing import MinMaxScaler  # StandardScaler
```

```python
scaler_mm = MinMaxScaler()
```

```python


penguins[['mmLen', 'mmDep']] = scaler_mm.fit_transform(penguins[['bill_length_mm', 'bill_depth_mm']])
```

```python
_, axs = plt.subplots(nrows=2)
sns.lineplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', ax=axs[0])
sns.lineplot(data=penguins, x='mmLen', y='mmDep', ax=axs[1]);
```

```python
penguins[['mmMass']] = scaler_mm.fit_transform(penguins[['body_mass_g']])
penguins.head()
```

```python
row_count = penguins.shape[0]
ns = np.arange(0, row_count)
_, axs = plt.subplots(nrows=2)
axs
sns.lineplot(data=penguins, x=ns, y='body_mass_g', ax=axs[0])
sns.lineplot(data=penguins, x=ns, y='mmMass', ax=axs[1]);
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc = OneHotEncoder()
res = enc.fit_transform(penguins[['species']])
enc.categories_
res.toarray()
penguins[enc.categories_[0]] = res.toarray()
penguins.drop(columns=['species'], inplace=True)
penguins
```

```python
res = enc.fit_transform(penguins[['island']])
enc.categories_
res.toarray()
penguins[enc.categories_[0]] = res.toarray()
penguins.drop(columns=['island'], inplace=True)
penguins
```

```python
res = enc.fit_transform(penguins[['sex']])
enc.categories_
res.toarray()
penguins[enc.categories_[0]] = res.toarray()
penguins.drop(columns=['sex'], inplace=True)
penguins
```

```python
from sklearn.decomposition import PCA
```

```python

pca = PCA(n_components=4)
penguins_pca = pca.fit_transform(penguins)
print('transformed penguins', penguins_pca[:3], '', sep='\n')
print('explained_variance', pca.explained_variance_)
print('explained_variance_ratio_', pca.explained_variance_ratio_)
```
